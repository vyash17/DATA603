```python
from spellchecker import SpellChecker
from functools import reduce
from collections import defaultdict
```

```python
# Create English dictionary
#birthdate 13th april 2001
spell_checker = SpellChecker(language='en')
```

```python
# Mapper function
def mapper(line):
    counts = defaultdict(int)
    for word in line.strip().split():
        correction = spell_checker.correction(word)
        if correction != word:
            counts[correction] += 1
    return counts.items()
```

```python
# Reducer function
def reducer(counts1, counts2):
    combined_counts = defaultdict(int)
    for word, count in counts1:
        combined_counts[word] += count
    for word, count in counts2:
        combined_counts[word] += count
    return combined_counts.items()
```

```python
# Read text file
with open('file2.txt', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
# MapReduce
non_english_word_counts = reduce(reducer, map(mapper, lines))
```

```python
# Print non-English word counts
for word, count in non_english_word_counts:
    print(word, count)
```
