```python
from functools import reduce
from collections import defaultdict
```

```python
# Mapper function
#birthdate 13th april 2001
def mapper(line):
    counts = defaultdict(int)
    for word in line.strip().split():
        counts[word] += 1
    return counts.items()
```

```python
# Reducer function
def reducer(counts1, counts2):
    combined_counts = defaultdict(int)
    for word, count in counts1:
        combined_counts[word] += count
    for word, count in counts2:
        combined_counts[word] += count
    return combined_counts.items()
```

```python
# Read text file
with open('file1.txt', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
# MapReduce
word_counts = reduce(reducer, map(mapper, lines))
```

```python
# Print word counts
for word, count in word_counts:
    print(word, count)
```
